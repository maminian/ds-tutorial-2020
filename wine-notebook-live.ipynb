```python
import numpy as np
from matplotlib import pyplot
import pandas

import utils
```

```python
df = utils.load_wine()
```

```python
data = df.iloc[:,:11].values
labels = df['type'].values
```

```python
labels
```

```python
ec = {wine:np.where(labels==wine) 
      for wine in ['red', 'white'] 
     }
```

```python
for k,v in ec.items():
    pyplot.scatter(data[v,0], data[v,1], label=k)
pyplot.legend()
```

```python
fig,ax = pyplot.subplots( 11,1, figsize=(6,22) )

for j in range(11):
    # for each chemical property,
    for k,v in ec.items():
        # for each type of wine, and a given chemical property, make histograms.
        ax[j].hist(data[v,j], label=k)
        
```
